```python
from ImageFiltering import Filters
import numpy as np
```

```python
# Assinging the class with image path  ->  Hint: Filters(Image Path)
filters = Filters(r'C:\Users\Zahran Zahid\CvLabWork\DIP3E_Original_Images_CH02\Fig0222(b)(cameraman).tif')
```

##### **Applying Gaussian filter, Mean filter, Average filter & Linear filter**

```python
g_filter= filters.gaussian_filter(5, 5, 2)

m_filter= filters.mean_filter(5)

avg_filter= filters.average_filter(5)
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1, 8, -1],
                                  [-1, -1, -1]])

lr_filter = filters.linear_filter(edge_detection_kernel)
```

##### **Ploting the Results**

```python
filters.show_images(filters.image, g_filter, m_filter,
                    titles=['Original', 'Gaussian Filter', 'Mean Filter'])
```

```python
filters.show_images(filters.image, avg_filter, lr_filter,
                    titles=['Original', 'Average Filter', 'Linear Filter'])
```
